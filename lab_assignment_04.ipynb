Please conduct the following analysis:

(2)  Visualize the count for each value in the 'Study Hours' column. Note: Study Hours values comes from this >= 6 hours Good >= 4 hours Average < 2 hours Poor.

(3) Using factorplot to show the relationship between numerical and categorical value. Please use figure to show relation between [Study Hours] and [End Semester Percentage]. Percentage is calculated based on the following:

if percentage >=80 then Best
If percentage >= 60 but less than 80 then Very Good
If percentage >= 45 but less than 60 then Good
If Percentage >= 30 but less than 45 then Pass
If Percentage < 30 then Fail

(4) Get the correlation from the data, then plug it to heatmap function and show it as following:



```python
# Your answer here (code + explanation):
#Importing Libraries
import pandas as pd

#Reading Data and printing top 5 rows in data
df = pd.read_csv('C:/Users/Sasi9/Downloads/INFO5502-Summer2022-main/INFO5502-Summer2022-main/datasets/lab-assignment-04-student-performance.csv')
df.head()
```

```python
#Checking no of rows and columns in Data
df.shape
```

```python
#1) Description of each column using Describe() function
print(df.describe())
```

```python
#Printing No of rows, no of Unique elements and Frequency of each element in column
for i in df.columns:
    print("No of rows in",i,"is:",df[i].shape[0])
    print("No of Uniques elements in",i,":",df[i].nunique())
    print("Frequency of Each Unique element in",i)
    print(df[i].value_counts())
```

```python
#2) Visualizing count plot for Study Hour 
import seaborn as sns
import matplotlib.pyplot as plt
ax = sns.countplot(x="sh", data=df)
ax.set(xlabel='Study Hours', ylabel='Count')
ax.bar_label(ax.containers[0])
plt.show()
```

```python
#3) COnverting Percentage column into numeric and Ploting Categorical Vs Numerical Column 
def New_col_Creation(twp):
    if twp == "Best":
        return 80
    elif twp == "Vg":
        return 60
    elif twp == "Good":
        return 45
    elif twp == "Pass":
        return 30
    else:
        return 20

df["Num_Perc"] = df["twp"].apply(lambda twp: New_col_Creation(twp))
df["Num_Perc"].value_counts()
```

```python
ax = sns.factorplot(x='sh',y='Num_Perc', kind = 'strip',data=df)
ax.set(xlabel='Study Hours', ylabel='Percentage')
plt.show()
```

```python
#we are dropping ms col because it don't have any variance, 'ms' is constant column with 1 element
# we are dropping newly created column
df.drop(['Num_Perc','ms'],axis=1,inplace=True)
```

```python
#4) Correlation Plot
#Converting all columns into categorical and encoding all the variables into numerical 
for i in df.columns:
    df[i] = df[i].astype('category')
    df[i] = df[i].cat.codes

#Creating correlation matrix by using .corr() function and visualizing using sns.heatmap
corr = df.corr()
f = plt.figure(figsize=(19, 15))
sns.heatmap(corr,annot=True,cmap="Oranges")
plt.show()
```
